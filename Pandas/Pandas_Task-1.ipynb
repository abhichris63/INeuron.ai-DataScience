```python
# 1. Find out how many male and female passenger was onboarded.

male = df[df["Sex"][0:]=="male"]
male["Sex"].count() ## 577 Males
```

```python
female = df[df["Sex"][0:]=="female"]
female["Sex"].count() ## 314 Females
```

```python
# 2. how many survived we have 

survived = df[df['Survived'] == 1]
survived['Survived'].count() ## 342 are Survived
```

```python
# 3. how many casuality we have 

Casuality = df[df["Survived"]==0]
Casuality["Survived"].count() ## 549 Casuality are there
```

```python
# 4. what is name of a person who is the eldest one.

df[df["Age"]==max(df["Age"])]["Name"] ## Barkworth, Mr. Algernon Henry Wilson is the Eldest Person.
```

```python
# 5. how many passenger do we have in first, second and third class.

Passenger_class1 = df[df["Pclass"]==1]
print("In First Class : ",Passenger_class1['Pclass'].count()) # Passenger Class First is 216

Passenger_class2 = df[df["Pclass"]==2]
print("In Second Class : ",Passenger_class2['Pclass'].count()) # Passenger Class second is 184

Passenger_class3 = df[df['Pclass']==3]
print("In Third Class : ",Passenger_class3["Pclass"].count())  # Passenger Class Third is 491
```

```python
# 6. how many person we have whose name start with 's'

count=0
for i in df["Name"]:
    if i[0]=="S":
        count=count+1
        print(i)
print(f"There are {count} Names which starts with 's'") ## There are 86 Names which starts with 's'
```

```python
# 7. try to create a new column which is a summation of "SibSp" and "parch"

df["Sum"] = df["SibSp"] + df["Parch"] # New Colmun Sm is created and added the sum of column 'SibSp' & 'Parch'
df
```

```python
# 8. how many person do we have below age of 25.

persons = df[df["Age"]<25]
persons["Age"].count() ## There are 278 Persons are below age 25 years. 
```

```python
# 9. how many person died whose age was less than 40.

age = df[df["Age"]<40]
dead=age["Survived"]==0
a = age[dead].count()
a["Survived"] #Person who Died whose age less than 40 are : 322 persons
```

```python
# 10. from cabin column seperate text and numeric value.

df["Cabin_Letter"] = df["Cabin"].str.extract('([A-Za-z]+)')
df['Cabin_Number'] = df.Cabin.str.extract('(\d)', expand=True)
df
```
